```python
inst = 1
field = 'SWIRE'
lphdir = '../../lephare/lephare_dev/panstarrs/'
catcoorddir='../../doc/20170617_Stacking/maps/catcoord/TM'+str(inst)+'/'
hdr = ['ra', 'dec', 'x', 'y','gMeanPSFMag', 'rMeanPSFMag', \
               'iMeanPSFMag', 'zMeanPSFMag', 'yMeanPSFMag','dbscanClass']
fname=catcoorddir+'PanSTARRS/'+ field + '.txt'
df = pd.read_csv(fname)
```

```python
df.head()
```

```python
data = np.loadtxt(lphdir + field + '_WISE.out')
```

```python
# write lephare output to df
 
nband_use = np.array([data[:,12]-2,data[:,12]-3,data[:,12]-3]).T
# star, gal, qso
chi = np.array(data[:,[20,5,17]])
chi = chi / nband_use

# 3 type classification
# 0 - star, 1 - gal, 2 - qso
lph_class3 = np.argmin(chi,axis=1)
# -1 - star, 1 - gal, 2 - qso
lph_class3[lph_class3==0] = -1
# set dof = 0 src to -999
lph_class3[nband_use[:,1]<=0] = -999

# 2 type classification
# 0 - star, 1 - gal
lph_class = np.argmin(chi[:,:2],axis=1)
# -1 - star, 1 - gal
lph_class[lph_class==0] = -1
# set dof = 0 src to 3
lph_class[nband_use[:,1]<=0] = -999


df['z_gal'] = data[:,1]
df['z_qso'] = data[:,16]
df['dof_gal'] = np.array(nband_use[:,1], dtype=int)
df['chi2_star'] = data[:,20]
df['chi2_gal'] = data[:,5]
df['chi2_qso'] = data[:,17]
df['lphClass3'] = lph_class3
df['lphClass'] = lph_class
```
