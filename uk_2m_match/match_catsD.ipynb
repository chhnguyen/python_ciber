# read in UKIDSS and 2MASS catcoord data

```python
inst=1
field='SWIRE'
quad='D'

catcoorddir='../doc/20170617_Stacking/maps/catcoord/TM'+str(inst)+'/'

ukdfname = list(pd.read_csv(catcoorddir+'UKIDSS/'+field+'_'+quad+'_uk.txt',header=0,sep='|'))
ukdf = pd.read_csv(catcoorddir+'UKIDSS/'+field+'_'+quad+'_uk.txt',header=0,sep=',',names=ukdfname)

tmdfname = list(pd.read_csv(catcoorddir+'PSC/'+field+'_'+quad+'_2m.txt',header=0,sep='|'))
tmdf = pd.read_csv(catcoorddir+'PSC/'+field+'_'+quad+'_2m.txt',header=0,sep=',',names=tmdfname)
```

# match 2MASS counterparts in UKIDSS

```python
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.coordinates import match_coordinates_sky

coordtm = SkyCoord(ra=tmdf['ra']*u.degree, dec=tmdf['dec']*u.degree, frame='icrs')
coorduk = SkyCoord(ra=ukdf['ra']*u.degree, dec=ukdf['dec']*u.degree, frame='icrs')

idx, d2d,_ = match_coordinates_sky(coordtm, coorduk)
```

```python
unique, counts = np.unique(idx, return_counts=True)
num,c = np.unique(counts, return_counts=True)


print('match %d 2MASS sources in %s quad %s'%(len(idx),field,quad))
print('%d/%d (%.3f%%) 2MASS sources has unique counter part in UKIDSS'%(c[0],len(idx),c[0]/len(idx)*100))

print(num,c)
```

```python
ukdf_match=ukdf.iloc[idx].copy()
ukdf_match=ukdf_match.reset_index(drop=True)
ukdf_match.head()
```

# Create new data frame for info from both catalogs

```python
newtm=tmdf[['idx','x','y','mAB_j','mAB_h','mAB_k']].copy()
newtm=newtm.rename(index=str,columns={'idx':'tm_idx','x':'tm_x','y':'tm_y',\
                                      'mAB_j':'tm_mAB_j','mAB_h':'tm_mAB_h','mAB_k':'tm_mAB_k'})
newuk=ukdf_match[['idx','x','y','mAB_j','mAB_k','mergedClass','pstar','pgalaxy','pnoise']].copy()
newuk=newuk.rename(index=str,columns={'idx':'uk_idx','x':'uk_x','y':'uk_y','mAB_j':'uk_mAB_j','mAB_k':'uk_mAB_k'})
new=pd.concat([newtm, newuk], axis=1)

new['dmatch_arcsec']=d2d.arcsec
new['dm_j']=new['tm_mAB_j']-new['uk_mAB_j']
new['dm_k']=new['tm_mAB_k']-new['uk_mAB_k']

new=new[['tm_idx','uk_idx','tm_x','uk_x','tm_y','uk_y','dmatch_arcsec',\
         'tm_mAB_j','uk_mAB_j','dm_j','tm_mAB_h','tm_mAB_k','uk_mAB_k','dm_k',\
        'mergedClass','pstar','pgalaxy','pnoise']]
```

# clean up new data frame

```python
mask =  (new.tm_x > -0.5) & (new.tm_x < 511.5) & (new.tm_y > -0.5) & (new.tm_y < 511.5)
mask = -mask & (new.dmatch_arcsec>1)

new1=new.copy()
new1.loc[mask,'mergedClass'] = -2

mask1 = (new.tm_y > 468) & (new.tm_x > 288)
new1.loc[mask1,'mergedClass'] = -2
```

```python
groups = new1.groupby('mergedClass')

# Plot
fig, ax = plt.subplots(figsize=(10,8))
namedict={1:'gals',-1:'stars',-2:'others',-3:'others'}
cdict={1:'b',-1:'r',-2:'g',-3:'g'}
for name, group in groups:
    ax.plot(group.tm_x, group.tm_y, marker='o', linestyle='', ms=1, label=namedict[name],c=cdict[name])
ax.legend()
ax.axhline(0)
ax.axvline(0)
ax.axhline(512)
ax.axvline(512)
```
